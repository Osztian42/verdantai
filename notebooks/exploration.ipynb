```python
import pandas as pd

df = pd.read_csv("../data/crop_yield_dataset.csv")

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
zero_yield = df[df["Crop_Yield"] == 0]

print(zero_yield.head())

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()

plt.hist(df["Crop_Yield"], bins=30)
plt.xlabel("Crop Yield")
plt.ylabel("Frequency")
plt.title("Distribution of Crop Yield")
plt.show()

plt.scatter(df["Temperature"], df["Crop_Yield"])
plt.xlabel("Temperature")
plt.ylabel("Crop Yield")
plt.title("Temperature vs Crop Yield")
plt.show()

sns.scatterplot(x="Temperature", y="Crop_Yield", data=df)
plt.title("Temperature vs Crop Yield")
plt.show()

sns.scatterplot(x="Humidity", y="Crop_Yield", data=df)
plt.title("Humidity vs Crop Yield")
plt.show()

sns.boxplot(x="Crop_Type", y="Crop_Yield", data=df)
plt.xticks(rotation=45)
plt.show()


sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 7))

sns.boxplot(
    data=df,
    x="Crop_Type",
    y="Crop_Yield",
    palette="Greens"
)

plt.title(
    "Crop Yield Distribution by Crop Type",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Crop Type", fontsize=12)
plt.ylabel("Crop Yield", fontsize=12)
plt.xticks(rotation=35)
plt.tight_layout()
plt.show()

# Theme
plt.style.use("dark_background")

# Create figure
fig, ax = plt.subplots(figsize=(14, 8))

# Background colors
fig.patch.set_facecolor("#07140F")
ax.set_facecolor("#0D1F17")

# Plot
sns.violinplot(
    data=df,
    x="Crop_Type",
    y="Crop_Yield",
    palette=["#2E5E3B", "#4C8C4A", "#7BC96F", "#A4DE8E"],
    linewidth=2,
    inner="quartile",
    ax=ax
)

# Titles
ax.set_title(
    "Crop Yield Distribution Across Plant Types",
    fontsize=24,
    fontweight="bold",
    color="#E8F5E9",
    pad=25
)

ax.set_xlabel(
    "Crop Type",
    fontsize=14,
    color="#C8E6C9",
    labelpad=15
)

ax.set_ylabel(
    "Crop Yield",
    fontsize=14,
    color="#C8E6C9",
    labelpad=15
)

# Tick styling
ax.tick_params(colors="#A5D6A7", labelsize=12)

# Grid
ax.grid(
    color="#1F3B2D",
    linestyle="--",
    linewidth=0.5,
    alpha=0.6
)

# Remove top/right borders
sns.despine()

plt.xticks(rotation=25)

plt.tight_layout()
plt.show()
```
